```julia
using CSV
using DataFrames; df = CSV.read(file, DataFrame)
```

```julia
newtreeg = []
for guard in df[!, :TreeGuard]
    if ismissing(guard)
        guard = "None"
    end
    push!(newtreeg, guard)
end
```

```julia
using Lathe.preprocess: OrdinalEncoder
```

```julia
df[!, :TGs] = encoder.predict(df[!, :TreeGuard])
```

```julia
using Lathe.stats: OneTailed, T_Dist
```

```julia
df[!, :scaled_ar] = scaler.predict(df[!, :Area])
```
